#### Import SpaCy and load medium-sized language model

```python
import spacy
nlp = spacy.load('en_core_web_md')
```

#### Create the recommendation function

```python
def recommend_sim_movie(input_descrip):
    input_doc = nlp(input_descrip)
    max_similarity = -1
    best_match_index = -1

    for i, doc in enumerate(docs):
        similarity = input_doc.similarity(doc)
        if actual_titles[i] != input_descrip and similarity > max_similarity:
            max_similarity = similarity
            best_match_index = i

    return actual_titles[best_match_index]
```

#### Read in the file and prepare text

```python
movies_labels = []
movies_descrip = []

with open(
    'movies.txt', 'r', encoding='utf-8') as file:
    for line in file:
        if ':' in line:
            label, description = line.strip().split(':', 1)
            movies_labels.append(label.strip())        
            movies_descrip.append(description.strip())

actual_titles = [
    "How to Train Your Dragon",
    "Reign of the Supermen",
    "Suspiria",
    "Holmes & Watson",
    "Summer ’03",
    "The Captain",
    "The Last Boy",
    "A Star is Born",
    "The Most Wonderful Time of the Year",
    "Ladies in Black",
    "Planet Hulk"
]
```

#### Add movie titles

```python
movies_titles = actual_titles[:len(movies_labels)]
```

#### Convert descriptions into SpaCy Doc objects

```python
docs = [nlp(desc) for desc in movies_descrip]
```

#### Apply movie recommendation function

```python
planet_hulk_descrip = """
    Will he save their world or destroy it? 
    When the Hulk becomes too dangerous for the Earth, 
    the Illuminati trick Hulk into a shuttle and launch 
    him into space to a planet where the Hulk can live 
    in peace. Unfortunately, Hulk lands on the planet 
    Sakaar where he is sold into slavery and trained as 
    a gladiator.
    """

recommended_movie = recommend_sim_movie(planet_hulk_descrip)

print(f"Would you like to watch this movie next: '{recommended_movie}'?")
```
