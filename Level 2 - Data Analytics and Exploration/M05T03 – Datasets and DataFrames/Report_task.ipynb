### Datasets and DataFrames Task

Import libraries

```python
import pandas as pd
```

Create a pandas dataframe

```python
df = pd.read_csv('balance.txt',sep=' ')
```

Write the code needed to produce a report that provides the following information:
* Compare the average income based on ethnicity.
* On average, do married or single people have a higher balance?
* What is the highest income in our dataset?
* What is the lowest income in our dataset?
* How many cards do we have recorded in our dataset? (Hint: use `sum()`)
* How many females do we have information for vs how many males? (Hint: use `count()`).  

For a list of all methods for computation of descriptive stats, explore the pandas documentation

(https://pandas.pydata.org/pandas-docs/stable/reference/frame.html#computations-descriptive-stats)

Compare the average income based on ethnicity.

```python
# Get average income
ave_income_by_ethnicity = df.groupby('Ethnicity')['Income'].mean()
display(ave_income_by_ethnicity)

# Sort from highest to lowest
sort_income_ethnicity = (
    df.groupby('Ethnicity')['Income'].mean().
    sort_values(ascending=False)
    )
ethnicity_sorted = sort_income_ethnicity.index.tolist()

# Print the answer
print(f"{ethnicity_sorted[0]} people have the highest average income.\n"
      f"{ethnicity_sorted[1]} people have the second-highest average income.\n"
      f"{ethnicity_sorted[2]} people have the lowest average income.\n")
```

On average, do married or single people have a higher balance?

```python
# Display data
ave_bal_by_marital_status = df.groupby('Married')['Balance'].mean()
display(ave_bal_by_marital_status)

# Answer the question directly
answer = (
    "married" 
    if ave_bal_by_marital_status['Yes'] > ave_bal_by_marital_status['No']
    else "single"
)
print(f"Conclusion: People who are {answer} have a higher bank balance.")
```

What is the highest income in our dataset?

```python
highest_income = df['Income'].max()
print(f'The highest income is: {highest_income:.2f}.')
```

What is the lowest income in our dataset?

```python
lowest_income = df['Income'].min()
print(f'The lowest income is: {lowest_income:.2f}.')
```

How many cards do we have recorded in our dataset? (Hint: use `sum()`)

```python
tot_cards = df['Cards'].sum()
print(f'Total number of bank cards on record: {tot_cards}.')
```

How many females do we have information for vs how many males? (Hint: use `count()`).  

```python
gender_count = df['Gender'].value_counts()

print(
    f"There are {gender_count['Female']} females and " 
    f"{gender_count['Male']} males in the dataset."
)
```
