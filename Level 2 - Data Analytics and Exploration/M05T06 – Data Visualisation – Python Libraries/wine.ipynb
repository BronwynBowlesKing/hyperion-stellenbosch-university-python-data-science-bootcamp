Import libraries

```python
import matplotlib.pyplot as plt
import pandas as pd
```


Read in the wine.csv file, which is a sample of wine reviews collected in 2017.


```python
wine_dataset = pd.read_csv('wine.csv')
```

Use the variety column to filter the dataset to only contain “Cabernet Sauvignon”, “Pinot Noir”, and “Chardonnay” wines using the pandas isin() function.

```python
target_wine = wine_dataset[wine_dataset['variety'].isin(['Cabernet Sauvignon', 'Pinot Noir', 'Chardonnay'])]
target_wine = target_wine.sort_values(by='variety')
display(target_wine)
```

Create a multi-plot grid with a histogram plot of the “points” column for each of these three wine varieties.

```python
varieties = ['Cabernet Sauvignon', 'Pinot Noir', 'Chardonnay']
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

for ax, variety in zip(axes, varieties):
    data = target_wine[target_wine['variety'] == variety]['points']
    ax.hist(data, bins=10, color='darkred', edgecolor='black')
    ax.set_title(f'{variety} points distribution')
    ax.set_xlabel('Award points')
    ax.set_ylabel('Wine count')

plt.show()
```
